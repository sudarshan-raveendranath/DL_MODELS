```python
#also we can use !wget for downlad the image directly from the web
# we can use matplotlib.image or pillow or cv2 for image processing
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image #for image processing (Pillow)
import cv2 #for image processing (OpenCV)
```

```python
#loading the image through matplotlib.image

img = mpimg.imread('puppy.jpeg')
```

```python
type(img) #numpy.ndarray
```

```python
img.shape # 2868 means 2868 pixels in height and 4302 pixels in width and 3 means 3 channels RGB. if that is grayscale image then it will be 2868,4302,1
```

```python
# total number of pixels in the image = 2868*4302 = 12345736
# we have 3 matrices of 2868*4302 each for RGB channels. one for red channel, one for green channel and one for blue channel
```

```python
print(img)
```

```python
#convert numpy array to image
img_plot = plt.imshow(img)
plt.show()
```

```python
#if we are giving more images to a neural network or something, all the images should be of same size. so we need to resize the images
# resizing the image using pillow
img = Image.open('puppy.jpeg')
img_resized = img.resize((200,200))
```

```python
print(img_resized.size) # (200,200)
```

```python
img_resized.save('puppy_resized.jpeg') #saving the resized image
```

```python
img_res = mpimg.imread('puppy_resized.jpeg') #loading the resized image
img_res_plot = plt.imshow(img_res) #plotting the resized image
plt.show()
```

```python
#convert color image to grayscale image using opencv
img = cv2.imread('puppy.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)  #converting rgb to gray
```

```python
img_gray_plot = plt.imshow(grayscale_img, cmap='gray')
plt.show()
```

```python
grayscale_img.shape #clr channel is 1 bcz there is only one channel
```

```python
cv2.imwrite('puppy_grayscale_img.jpeg', grayscale_img)
```
